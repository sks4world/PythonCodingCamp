```python
def print_col_types():
    cols=df.columns
    for col in cols:
        print('Column ' + col + ' is of data type '+ str(type(df[col])))
        str1=str(type(df[col]))
        str2=str1[7:]
        str3=str2[:-1]
        #print('Column ' + col + ' is of data type '+ str3 )
        
        
def main():
    print_col_types()
    print(type(df))
    
main()
```

```python
print(type(df))
df.columns=['prog_lang', 'avg_sal', 'sal_inc_pc']
df['avg_sal']=df['avg_sal'].str.replace('$','').str.replace(',','')
df['sal_inc_pc'].fillna(0.0, inplace=True)
#df['sal_inc_pc']=df['sal_inc_pc'].fillna(method='ffill', inplace=True)
```

```python
df.head(10)
```

```python
df = df.astype({'avg_sal':'int', 'sal_inc_pc':'float'})
#df = df.astype({'avg_sal':'int', 'sal_inc_pc':'int'})
```

```python
print_col_types()
```

```python
df.prog_lang=df.prog_lang.str.replace('\d+','').str.replace('.','')
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (14,5)

plt.hist(df.avg_sal)
plt.show()
```

```python
import numpy as np

height = df.avg_sal
bars = df.prog_lang
y_pos = np.arange(len(bars))
 
# Create bars
plt.bar(y_pos, height, 0.3)
 
# Create names on the x-axis
plt.xticks(y_pos, bars)
plt.title('Salary by Programming language')

 
# Show graphic
plt.show()
 
```

```python
height1 = df.sal_inc_pc
bars1 = df.prog_lang
y_pos1 = np.arange(len(bars))
 
# Create bars
plt.bar(y_pos1, height1, 0.3)
 
# Create names on the x-axis
plt.xticks(y_pos1, bars1)
plt.grid(color='black', linestyle='-', linewidth=0.2)
plt.title('Percentage change YoY in Salary')

 
# Show graphic
plt.show()
```
